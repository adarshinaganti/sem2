# Exercise 1

````python
df = pd.read_csv('mtcars.csv')
df.head()
````

Create cross tabulation using pandas. Cross-tabulation features will help find occurrences for the combination of values for two columns. Tabulate the cylinders and gears cars.

````python
pd.crosstab(df['cyl'],df['gear']).head()
````

Heatmap of the crosstab result. The best way to visualize a cross tab is to use a heatmap.

````python
sns.heatmap(pd.crosstab(df['cyl'],df['gear']).head(), annot=True, fmt='d', cmap='YlGnBu')
plt.title('Crosstab Heatmap: Cylinders vs Gears')
plt.xlabel('Gears')
plt.ylabel('Cylinders')
plt.gca().collections[0].colorbar.set_label('Frequency')
````

```sort_values()``` takes the column names, based on which the records need to be sorted. By default, sorting is done in ascending order.

````python
df[['model','mpg']].sort_values('mpg',ascending=False).head()
````

```groupby``` and aggregate: Average MPG by Number of Cylinders.

````python
df.groupby('cyl')['mpg'].mean().head()
````

Bar chart is a frequency chart for qualitative variable (or categorical variable). Bar chart can be used to assess the most-occurring and least-occurring categories within a dataset.

````python
sns.barplot(x ='cyl', y ='mpg', data = df, ci=None)
````

Scatter plot of mpg vs. hp

````python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='hp', y='mpg', data=df)
plt.title("MPG vs HP")
plt.show()
````

Histogram of mpg

````python
plt.figure(figsize=(6, 4))
sns.histplot(df['mpg'], bins=10, kde=True)
plt.title("Distribution of MPG")
plt.show()
````

Create side-by-side plots for mpg

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(df['mpg'], bins=20, kde=True, ax=axes[0], color='blue')
axes[0].set_title('MPG Distribution')
axes[0].set_xlabel('Miles Per Gallon (MPG)')
axes[0].set_ylabel('Count')

sns.boxplot(x='cyl', y='mpg', data=df, ax=axes[1], palette='viridis')
axes[1].set_title('MPG by Number of Cylinders')
axes[1].set_xlabel('Number of Cylinders')
axes[1].set_ylabel('Miles Per Gallon (MPG)')

plt.show()
````

Pair plot of numerical columns

````python
sns.pairplot(df[['mpg', 'hp', 'wt', 'qsec']])
plt.show()
````

Violin plot of mpg by cylinder count

````python
plt.figure(figsize=(6, 4))
sns.violinplot(x='cyl', y='mpg', data=df)
plt.title("MPG Distribution by Cylinder Count")
plt.show()
````

Regression plot of mpg vs. weight

````python
plt.figure(figsize=(6, 4))
sns.regplot(x='wt', y='mpg', data=df)
plt.title("MPG vs Weight (Regression)")
plt.show()
````

Swarm plot of mpg by cylinder count

````python
plt.figure(figsize=(6, 4))
sns.swarmplot(x='cyl', y='mpg', data=df)
plt.title("MPG by Cylinder Count (Swarm Plot)")
plt.show()
````

# Exercise 2

````python
df = pd.read_csv('deliveries.csv')
df.head()
````

````python
# Cleaning data
df.replace('Royal Challengers Bangalore', 'Royal Challengers Bengaluru', inplace=True)
df.replace('Delhi Daredevils', 'Delhi Capitals', inplace=True)
df.replace('Rising Pune Supergiant', 'Rising Pune Supergiants', inplace=True)
df.replace('Kings XI Punjab', 'Punjab Kings', inplace=True)
df.replace('Gujarat Lions', 'Gujarat Titans', inplace=True)
df.head()
````

Find and visualize the total runs scored by each team.

````python
df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
````

````python
plt.figure(figsize=(12, 8))
bar = sns.barplot(x=df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False).index, y=df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False).values, palette='viridis')
bar.bar_label(bar.containers[0])
plt.xlabel('Batting Team')
plt.ylabel('Total Runs Scored')
plt.title('Total Runs Scored by Each Team')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
````

Identify and display the names of the top 10 highest-scoring batters with their total runs scored.

````python
df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
````

````python
plt.figure(figsize=(12, 6))
bar = sns.barplot(x=df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10).index, y=df.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10).values, palette='coolwarm')
bar.bar_label(bar.containers[0])
plt.xlabel('Batter')
plt.ylabel('Total Runs Scored')
plt.title('Top 10 Highest-Scoring Batters')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
````

Find and display which of the top 3 overs has the highest run rate.

````python
(df.groupby(['match_id', 'inning'])['total_runs'].sum() / (df.groupby(['match_id', 'inning'])['ball'].count() / 6)).sort_values(ascending=False).head(3)
````

````python
top_inning_df = (df.groupby(['match_id', 'inning'])['total_runs'].sum() / (df.groupby(['match_id', 'inning'])['ball'].count() / 6)).sort_values(ascending=False).head(3).reset_index()
top_inning_df['Inning Label'] = top_inning_df['match_id'].astype(str) + ' - Inning ' + top_inning_df['inning'].astype(str)

plt.figure(figsize=(10, 6))
sns.barplot(x='Inning Label', y=0, data=top_inning_df, palette='viridis')
plt.xlabel('Match - Inning')
plt.ylabel('Run Rate')
plt.title('Top 3 Highest Run Rates Across All Innings')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
````
