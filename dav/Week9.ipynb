# Exercise 1

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

For mtcars dataset , create a categorical column based on mpg.

```python
mpg = df["mpg"]

# Define bins and labels
bins = [0, 18, 25, 40]
labels = ["low", "medium", "high"]

# Convert numbers to categorical
mpg_cat = pd.cut(mpg, bins=bins, labels=labels)
df['mpgCat'] = mpg_cat
df.head()
```

For mtcars dataset , Group by cars on fuel efficiency (mpg) and provide insights into them.

```python
df.groupby('mpgCat')['wt'].mean()
```

Box plot of weight for each MPG category.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='mpgCat', y='wt', data=df[['mpgCat','cyl', 'wt', 'mpg']])
plt.title('Weight by MPG Category')
plt.xlabel('MPG Category')
plt.ylabel('Weight')

plt.show()
```

### Commonly used string operations

```python
str1 = "  Data visualization is very interesting. I love Python Programming.   "

# Finding length
print(f'String length: {len(str1)}')

# Trim whitespace:
print(f'String 1: {str1.strip()}')

# Split a string into sentences
str2 = str1.split('.')
print(f'String 2: {str2}')

# Convert string into lower case (or upper case)
str3 = str1.lower()
print(f'String 3: {str3}')
```

```python
"python" in str3

# Index can be used to find a substring. index starts from 0.
print(str1.index('Python')) 

# Find can also be used, if substring does not exist in val, the method will return -1
print(str1.find(","))

# Replace can be used to replace substrings in string and assign to new string variable.
str2 = str1.replace('interesting','amazing')

# Count can be used to find number of exact matches of substring
print(str2.count('.'))
```

# Exercise 2

```python
df = pd.read_csv('sentiment.csv')
df.set_index('ID', inplace=True)
df.head()
```

In the sentiment feature and text, trim white spaces and convert to lowercase.

```python
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()
df['Text'] = df['Text'].str.strip().str.lower()

df[['Text', 'Sentiment']]
```

What are the sentiment categories? Tabulate and visualize the number of texts for each category.

```python
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()

table = pd.DataFrame(df['Sentiment'].value_counts()).reset_index()
table.columns = ['Sentiment Category', 'Number of Texts']

table.head()
```

```python
plt.figure(figsize=(20, 12))
sns.barplot(x='Sentiment Category', y='Number of Texts', data=table, palette='viridis')
plt.title('Number of Texts per Sentiment Category')
plt.xlabel('Sentiment Category')
plt.ylabel('Number of Texts')
plt.xticks(rotation=90, ha='right')

plt.show()
```

Tabulate and visualize the text length for each sentiment category.

```python
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()
df['Text Length'] = df['Text'].str.len()

df.groupby('Sentiment')['Text Length'].agg(['mean', 'median', 'min', 'max']).reset_index().head()
```

```python
plt.figure(figsize=(20, 12))
sns.barplot(x='Sentiment', y='Text Length', data=df.groupby('Sentiment')['Text Length'].mean().reset_index(), palette='coolwarm')
plt.title('Average Text Length per Sentiment Category')
plt.xlabel('Sentiment Category')
plt.ylabel('Average Text Length')
plt.xticks(rotation=90, ha='right')

plt.show()
```

Find and display the text(s) with the maximum number of likes.

```python
df['Likes'] = pd.to_numeric(df['Likes'])

for text in df[df['Likes'] == df['Likes'].max()]['Text']:
    print(f"- {text}")
```

Find and display the text(s) with the maximum reposts. What is its sentiment?

```python
df['Retweets'] = pd.to_numeric(df['Retweets'])

for index, row in df[df['Retweets'] == df['Retweets'].max()].iterrows():
    print(f"- Text: {row['Text']}")
    print(f"  Sentiment: {row['Sentiment']}")
```

Tabulate and visualize the top five countries that have posted the most texts with a positive sentiment.

```python
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()
df['Country'] = df['Country'].str.strip()

table = pd.DataFrame(df[df['Sentiment'] == 'positive']['Country'].value_counts().head(5)).reset_index()
table.columns = ['Country', 'Positive Texts']

table.head()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Country', y='Positive Texts', data=table, palette='magma')
plt.title('Top Five Countries with the Most Positive Texts')
plt.xlabel('Country')
plt.ylabel('Number of Positive Texts')

plt.show()
```

Find and display all the text(s) with the word ‘Travel’ in their hashtag.

```python
if not df[df['Hashtags'].str.contains('Travel', case=False, na=False)]['Text'].empty:
    for text in df[df['Hashtags'].str.contains('Travel', case=False, na=False)]['Text']:
        print(f"- {text}")
else:
    print("No texts found with 'Travel' in their hashtag.")
```

Find and display the most popular (most number of likes) posts (top 5) from the year 2020 which were of a negative sentiment.

```python
df['Likes'] = pd.to_numeric(df['Likes'])
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()

if not df[(df['Year'] == 2020) & (df['Sentiment'] == 'negative')].sort_values(by='Likes', ascending=False).head(5).empty:
    print("Top 5 negative sentiment posts with the most likes from the year 2020:")
    for index, row in df[(df['Year'] == 2020) & (df['Sentiment'] == 'negative')].sort_values(by='Likes', ascending=False).head(5).iterrows():
        print(f"- Text: {row['Text']}")
        print(f"  Likes: {row['Likes']}")
else:
    print("No negative sentiment posts found for the year 2020 in the provided data.")
```

Find and display a list of users and their posts whose texts are labeled ‘Happiness’ sentiment.

```python
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()
df['User'] = df['User'].str.strip()

if not df[df['Sentiment'] == 'happiness'].empty:
    for index, row in df[df['Sentiment'] == 'happiness'].iterrows():
        print(f"- User: {row['User']}, Post: {row['Text']}")
else:
    print("No posts found with 'Happiness' sentiment in the provided data.")
```

Tabulate and visualize the years and countrywide tweets of the sentiments. Pick tweets with sentiments, ‘Positive,’ ‘Negative,’ and ‘Neutral’ only.

```python
df['Sentiment'] = df['Sentiment'].str.strip().str.lower()
df['Platform'] = df['Platform'].str.strip().str.lower()

df[df['Platform'] == 'twitter'][df[df['Platform'] == 'twitter']['Sentiment'].isin(['positive', 'negative', 'neutral'])].groupby(['Year', 'Country', 'Sentiment']).size().reset_index(name='Count').pivot(index=['Year', 'Country'], columns='Sentiment', values='Count').fillna(0)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Country', y='Count', hue='Sentiment', data=df[df['Platform'] == 'twitter'][df[df['Platform'] == 'twitter']['Sentiment'].isin(['positive', 'negative', 'neutral'])].groupby(['Year', 'Country', 'Sentiment']).size().reset_index(name='Count'), dodge=True, palette='rocket')
plt.title('Country-wise Tweet Counts by Sentiment (2023)')
plt.xlabel('Country')
plt.ylabel('Number of Tweets')
plt.legend(title='Sentiment')

plt.tight_layout()
plt.show()
```
